**Matplotlib란**?
Matplotlib는 **파이썬(Python)**에서 가장 널리 사용되는 데이터 시각화 라이브러리입니다.

특히 2D 플롯(선 그래프, 막대 그래프, 산점도 등)을 만드는 데 많이 사용됩니다.

MATLAB과 비슷한 스타일로 설계되어 있어서 과학자나 엔지니어가 쉽게 적응할 수 있습니다.


⭐ 주요 특징
다양한 그래프 유형 지원 (선형, 막대, 파이 차트, 히트맵 등)

세밀한 커스터마이즈 가능 (글꼴, 색상, 레이아웃 등)

인터랙티브(마우스로 확대/축소 등) 기능 지원

다른 라이브러리(Pandas, NumPy 등)와 자연스럽게 연동





Matplotlib Line

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```

Matplotlib Labels and Title

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

Matplotlib Adding Grid Lines

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid()

plt.show()
```

Matplotlib Subplot

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

Matplotlib Bars

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```
